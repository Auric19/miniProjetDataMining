```python
import pandas as pd
pd.options.display.max_columns = 200
```

# PARTIE 1

Importer et afficher les CSV

```python
df_usagers = pd.read_csv("CSV/usagers_2016.csv")
print(df_usagers.shape)
df_usagers.head(10)
```

```python
df_lieux = pd.read_csv("CSV/lieux_2016.csv", low_memory=False)
print(df_lieux.shape)
df_lieux.head(10)
```

```python
df_vehicules = pd.read_csv("CSV/vehicules_2016.csv")
print(df_vehicules.shape)
df_vehicules.head(10)
```

```python
df_caracteristiques = pd.read_csv("CSV/caracteristiques_2016.csv")
print(df_caracteristiques.shape)
df_caracteristiques.head(10)
```

Changer les COM et DEP en chaine de caractères

```python
df_caracteristiques[['com','dep']] = df_caracteristiques[['com','dep']].astype(str)
print(df_caracteristiques.dtypes)
```

Fusionner les tableaux

```python
df_fusionA = df_usagers.merge(df_vehicules, on=['Num_Acc', 'num_veh'], how='outer')
df_fusionA.head(10)
```

```python
df_final = df_fusionB.merge(df_lieux, on='Num_Acc', how='outer')
df_final
```

# PARTIE 2

Proportion d’accidents entre 23h et 6h du matin par commune

```python
df_final["top_nuit"]=((df_final["hrmn"] > 2300) | (df_final["hrmn"] < 600)).astype(int)
df_final.top_nuit.mean()
```

Proportion d’accidents sur des autoroutes

```python
df_final["top_autoroute"] = (df_final.catr==1).astype(int)
df_final.top_autoroute.mean()
```

Proportion d’accidents en hiver

```python
df_final["top_hiver"]=(df_final["atm"]==4).astype(int)
df_final.top_hiver.mean()
```

Proportion d’accidents en agglomération

```python
df_final["top_agglo"]=(df_final["agg"]==2).astype(int)
df_final.top_agglo.mean()
```

Proportion d’accidents en conditions atmosphériques anormales

```python
df_final["top_anorm"]=(df_final["atm"]!=1).astype(int)
df_final.top_anorm.mean()
```

Séparer la colonne grav en 4 colonnes grâce au One-Hot Encoding

```python
one_hot_encoded = pd.get_dummies(df_final["grav"],prefix="grav")
df_final=pd.concat([df_final,one_hot_encoded],axis=1)

df_final["grav_1.0"]=df_final["grav_1.0"].astype(int)
df_final["grav_2.0"]=df_final["grav_2.0"].astype(int)
df_final["grav_3.0"]=df_final["grav_3.0"].astype(int)
df_final["grav_4.0"]=df_final["grav_4.0"].astype(int)
```

# Partie 6

```python
import numpy as np
import matplotlib.pyplot as plt
```

Préparation de la variable score

```python
regroupement_dep_com["score_grav"]=regroupement_dep_com["grav1"]+np.exp(1)*regroupement_dep_com["grav4"]+np.exp(2)*regroupement_dep_com["grav3"]+np.exp(3)*regroupement_dep_com["grav2"]
```

Application du boxplot sur la variable score

```python
plt.boxplot(regroupement_dep_com["score_grav"])
```

Transformation logarithmique

```python
regroupement_dep_com["score_norm"]=np.log(regroupement_dep_com.score_grav)
regroupement_dep_com.score_norm.hist()
```

Normalisation min max

```python
minmax_norm=(regroupement_dep_com["score_norm"] - regroupement_dep_com["score_norm"].min()) / ( regroupement_dep_com["score_norm"].max() - regroupement_dep_com["score_norm"].min())
minmax_norm
```

```python
regroupement_dep_com["score_norm"]=minmax_norm
```

```python
plt.boxplot(regroupement_dep_com["score_norm"])
```

Création d'une variable score_bin

```python
regroupement_dep_com["score_bin"]=""

score_bin_l=[]

for i in range(0,10815):
    if regroupement_dep_com["score_norm"][i]>regroupement_dep_com["score_norm"].mean():
      score_bin_l.append(1)
    else:
         score_bin_l.append(0)

regroupement_dep_com["score_bin"]=score_bin_l
```

Création d'une variable score_multi

```python
regroupement_dep_com["score_multi"]=""

score_multi_l=[]
for i in range(0,10815):
    if regroupement_dep_com["score_norm"][i]<regroupement_dep_com["score_norm"].max()/3:
      score_multi_l.append(1)
    elif regroupement_dep_com["score_norm"][i]>regroupement_dep_com["score_norm"].max()/3 and regroupement_dep_com["score_norm"][i]<(regroupement_dep_com["score_norm"].max()/3)*2:
         score_multi_l.append(2)
    else:
          score_multi_l.append(3)
            
regroupement_dep_com["score_multi"]=score_multi_l
```

```python
regroupement_dep_com.head(10)
```

Division du dataset en deux jeux train et test

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(regroupement_dep_com, test_size=0.2)
train, test = train.copy(), test.copy()
```
